# IBM Maximo Mobile


**Navigation**: The app features a navigator for quick access to various functions.

<img src="images/navigator.jpg" />

#### Completing a Work Order

1.  **Opening My Schedule**

    -   Navigate to the navigator and select "My Schedule" to view assigned work orders for the day.
    -   See a list of work orders categorized under "Assigned work."
<img src="images/workorderlist.jpg" />
      
2.  **Selecting a Work Order**

    -   Choose a preventive maintenance task for a centrifugal pump from the list.
    -   Click on "Get materials & tools" to view a checklist of required supplies (indicated by a blue button).

<img src="images/materialsandtools.jpg" />

3.  **Preparing Materials**

    -   Select checkboxes for required materials and tools.
    -   Confirm selection by clicking "Done."

<img src="images/materialsandtools_checklist.jpg" />

4.  **Navigating to the Asset**

    -   Follow map directions provided in the app to reach the centrifugal pump.
      
5.  **Starting Work**

    -   Return to the work order list and select "Start work" (indicated by a blue action).

<img src="images/startwork.jpg" />

6.  **Reviewing Work Order Details**

    -   Access detailed information about the asset, its location, attachments, and follow-up tasks on the work order details page.

<img src="images/workorderdetails.jpg" />

7.  **Scanning Asset Barcode**

    -   Scan the pump's barcode to confirm it's the correct asset.

<img src="images/pumpbarcode.jpg" />



<img src="images/afterbarcodescanning.jpg" />

8.  **Completing Task List**

    -   Review the detailed task list associated with the work order.
    -   Refer to safety information from attached manuals as needed.
    -   Mark each task complete as you finish.
  

<img src="images/tasklist.jpg" />


<img src="images/tasklistdetails.jpg" />


<img src="images/tasklistdetails_checked.jpg" />

9.  **Conducting Inspection**

    -   Open the inspection section within the work order.
    -   Review instructions from the supervisor (indicated by an 'i' icon).
    -   Start the inspection process (blue Start button).
    -   Answer required questions (indicated by blue dots).
    -   Capture and attach photos if issues (like a leaking seal) are identified during inspection.


<img src="images/inspections.jpg" />


<img src="images/inspections_photoattachment.jpg" />

10. **Finalizing Work**

    -   Complete all inspection questions and tasks.
    -   Mark inspection as complete and return to the main work order page.
    -   Click "Done" to finalize work order completion.


<img src="images/tasklistdetails_done.jpg" />

11. **Reporting Work**

    -   Report any issues or failures encountered during the inspection (e.g., leaking seal).
    -   Click "Complete work" to submit the completed work order.


<img src="images/failurereport1.png" />

<img src="images/pcr.jpg" />

<img src="images/failurereport2.jpg" />

12. **Completion**

    -   Return to the work list dashboard, ready for the next assigned task.
